# Files and Loops

### Opening and Reading Files

We open a file in Python using the open() function in which we write the name of the file and the "r", the mode for reading in files. More modes will be discussed later. Then we use the read() method in order to return a string representation of the text in a file.

```python
f=open("C:\\Users\\kopas\\Desktop\\instacart_example.csv","r")
data=f.read()
print(data)
```

### Splitting

Because the "data" file is only a string object that is not useful in Python, we use the <b>split()</b> method to turn a string object into a list of strings. Each line is separated by the string "\n", which is referred to as the new-line character. So we split our data on the character "\n".

```python
rows = data.split('\n')
print(rows[0:5])
```

### Loops

We now have a list representation of the dataset. Each element in the dataset is a string containing a comma (,) that separates the city name from the crime rate. Because they're strings, we can use the split() method on each of them to separate those values. In Python we do this with a <b>loop</b>. 

 for Loop

* Syntax - the words <b>for</b> and <b>in</b> need to be included in the statement
* Iterator variable - the variable name you decide to use to refer to each element in the list (city)
* Sequence - the variable you want to iterate over (cities)
* Colon - loop statements must end with a colon (:)

Loop Body

* Indentation - every line of code within the loop should be indented four spaces
* Logic - the actual code we want to execute for each element. In our example, we update the iterator variable (city) in each iteration of the loop.

```python
ten_rows = rows[0:10]
for row in ten_rows:
    print(row)
```

### List of Lists

How can we split the elements of the above list into more lists?

We start with the creation of an empty list <b>final_data</b>. Now we want to populate this list. Into a for loop we split the data on the comma delimiter and we append them into the empty list. 

```python
final_data = []
for row in rows:
    split_list = row.split(',')
    final_data.append(split_list)
print(final_data[0:5])
```

### Accessing Elements in a List of Lists

Let's say that we want to have access to the first value of the first list of final_list.  

```python
first_list = final_data[0]
first_list_first_value = first_list[0]
print(first_list_first_value)
```

Since using bracket notation once returns a list, you can use bracket notation again on the resulting list without having to assign it to a variable first. Using double bracket notation is simpler.

```python
first_list_first_value = final_data[0][0]
print(first_list_first_value)
```
